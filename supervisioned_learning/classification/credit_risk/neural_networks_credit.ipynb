```python
from sklearn.neural_network import MLPClassifier
```

```python
import pickle
with open('data/credit.pkl', 'rb') as f:
    x_training, y_training, x_test, y_test = pickle.load(f)
```

```python
(3+1)/2 # number of neurons per layer
```

```python
# sets epochs and tolerance (stopping criteria); 2 hidden layers, with 100 neurons
neural_network_credit  = MLPClassifier(max_iter=1500, verbose=True, tol=0.000100, 
                                       solver='adam', activation='relu', hidden_layer_sizes=(2, 2))
neural_network_credit.fit(x_training, y_training)
```

```python
predictions = neural_network_credit.predict(x_test)
predictions
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, classification_report
accuracy_score(y_test, predictions)
```

```python
import matplotlib.pyplot as plt
cm = confusion_matrix(y_test, predictions)
cm_disp = ConfusionMatrixDisplay(cm)
cm_disp.plot()
plt.show()
```

```python
print(classification_report(y_test, predictions))
```
