```python
import pickle
with open('data/credit.pkl', 'rb') as f:
    x_training, y_training, x_test, y_test = pickle.load(f)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_credit = LogisticRegression(random_state=1)
logistic_credit.fit(x_training, y_training)
```

```python
logistic_credit.intercept_ # b0 parameter
```

```python
# check parameters (age, income, loan)
logistic_credit.coef_
```

```python
predictions = logistic_credit.predict(x_test)
predictions
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, classification_report
accuracy_score(y_test, predictions)
```

```python
print(classification_report(y_test, predictions))
```

```python
import matplotlib.pyplot as plt
cm = confusion_matrix(y_test, predictions)
cm_disp = ConfusionMatrixDisplay(cm)
cm_disp.plot()
plt.show()
```
