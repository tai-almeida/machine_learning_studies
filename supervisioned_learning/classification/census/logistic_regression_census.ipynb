```python
import pickle
with open('data/census.pkl', 'rb') as f:
    x_training, y_training, x_test, y_test = pickle.load(f)
```

```python
from sklearn.linear_model import LogisticRegression

logistic_census = LogisticRegression(random_state=1)
logistic_census.fit(x_training, y_training)
```

```python
logistic_census.intercept_
```

```python
logistic_census.coef_
```

```python
predictions = logistic_census.predict(x_test)
predictions
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
accuracy_score(y_test, predictions)
```

```python
print(classification_report(y_test, predictions))
```

```python
import matplotlib.pyplot as plt
cm = confusion_matrix(y_test, predictions)
cm_disp = ConfusionMatrixDisplay(cm, display_labels=['<-50K', '>50K'])
cm_disp.plot()
plt.show()
```
