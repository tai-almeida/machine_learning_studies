```python
import pickle
with open('census.pkl', 'rb') as f:
    x_training, y_training, x_test, y_test = pickle.load(f)
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest_census = RandomForestClassifier(n_estimators=100, criterion='entropy', random_state=0)
random_forest_census.fit(x_training, y_training)
```

```python
predictions = random_forest_census.predict(x_test)
predictions
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, classification_report
accuracy_score(y_test, predictions)
```

```python
print(classification_report(y_test, predictions))
```

```python
random_forest_census.classes_
```

```python
import matplotlib.pyplot as plt
cm = confusion_matrix(y_test, predictions)
cm_disp = ConfusionMatrixDisplay(cm, display_labels=['<=50K', '>50K'])
cm_disp.plot()
plt.show()
```
